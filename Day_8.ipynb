```python
# Create a student dictionary and add first_name, last_name, gender, age, marital status, skills, country, city and address as keys for the dictionary
student = {'first_name': 'Wisdom', 'last_name': 'White', 'gender': 'Male', 'marital status': 'Single', 'skills': ['Data Science'], 'country': 'Canada', 'city': 'Toronto', 'address': '123 Maple Street'}
```

```python
# Get the length of the student dictionary
len(student)
```

```python
# Get the value of skills and check the data type, it should be a list
student['skills']
```

```python
# Modify the skills values by adding one or two skills
student['skills'].append('Research')

# For multiple skills
student['skills'].extend(['Analysis', 'Communication', 'Leadership'])

# To confirm
student['skills']
```

```python
# Get the dictionary keys as a list
student.keys()
```

```python
# Get the dictionary values as a list
student.values()
```

```python
# Change the dictionary to a list of tuples using items() method
student.items()
```

```python
# Delete one of the items in the dictionary
del student['marital status']
```
