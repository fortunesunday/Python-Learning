```python
# Convert the ages to a set and compare the length of the list and the set, which one is bigger?
age_set = set(age)
print(len(age))
print(len(age_set))

# the length of the list is bigger. This is so because there are duplicates and lists allow duplicates while sets do not.
```

```python
# Explain the difference between the following data types: string, list, tuple and set
# strings: This is a sequence of charcters
# Lists: Are a collection of ordered, mutable items
# Tuple: An ordered collection of items, they are usually nonmutable, and are indexed
# Sets: A collection of unordered items, usually unmutable.
```

```python
# 'I am a teacher and I love to inspire and teach people'. How many unique words have been used in the sentence? Use the split methods and set to get the unique words.
words = 'I am a teacher and I love to inspire and teach people'
split_words = words.split()
split_words

# Note that split_words is a list, therfore, convert to as set to get the unique words as sets do not allow duplicates.
words_set = set(split_words)

# Print 'words_Set'
print(words_set)

# To know the length of the unique words, use the len function
print(len(split_words))
print(len(words_set))

# There are 10 unique words.
```
